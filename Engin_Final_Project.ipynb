d. We learned from the central limit theorem that we can estimate the uncertainty on the mean
and standard deviation of a distribution (and indeed, any properties of the distribution) based
on how many times we sampled from that distribution.

i. You can test if your measured and expected variance are in statistical agreement by
estimating the uncertainty on your measured standard deviation from the central limit
theorem.

```python
print("Chem Library:")
print("Measured standard deviation:", np.std(chemLib["counts"]))
print("Measured variance:", (np.std(chemLib["counts"]))**2) # variance is standard deviation squared

len_chemLib = len(chemLib["counts"])
uncertainty_chemLib = np.sqrt(np.mean(chemLib["counts"])) / (np.sqrt(2 * np.sqrt(len_chemLib)))
print("Uncertainty on measured standard deviation:", uncertainty_chemLib)
print(" ")

print("Doe Library:")
print("Measured standard deviation:", np.std(doeLib["counts"]))
print("Measured variance:", (np.std(doeLib["counts"]))**2) # variance is standard deviation squared

len_doeLib = len(doeLib["counts"])
uncertainty_doeLib = np.sqrt(np.mean(doeLib["counts"])) / (np.sqrt(2 * np.sqrt(len_doeLib)))
print("Uncertainty on measured standard deviation:", uncertainty_doeLib)
print(" ")

print("Math Library:")
print("Measured standard deviation:", np.std(mathLib["counts"]))
print("Measured variance:", (np.std(mathLib["counts"]))**2) # variance is standard deviation squared

len_mathLib = len(mathLib["counts"])
uncertainty_mathLib = np.sqrt(np.mean(mathLib["counts"])) / (np.sqrt(2 * np.sqrt(len_mathLib)))
print("Uncertainty on measured standard deviation:", uncertainty_mathLib)
print(" ")

print("Music Library:")
print("Measured standard deviation:", np.std(musicLib["counts"]))
print("Measured variance:", (np.std(musicLib["counts"]))**2) # variance is standard deviation squared

len_musicLib = len(musicLib["counts"])
uncertainty_musicLib = np.sqrt(np.mean(musicLib["counts"])) / (np.sqrt(2 * np.sqrt(len_musicLib)))
print("Uncertainty on measured standard deviation:", uncertainty_musicLib)
print(" ")

print("Valley Life Sciences Building Library:")
print("Measured standard deviation:", np.std(vlsbLib["counts"]))
print("Measured variance:", (np.std(vlsbLib["counts"]))**2) # variance is standard deviation squared

len_vlsbLib = len(vlsbLib["counts"])
uncertainty_vlsbLib = np.sqrt(np.mean(vlsbLib["counts"])) / (np.sqrt(2 * np.sqrt(len_vlsbLib)))
print("Uncertainty on measured standard deviation:", uncertainty_vlsbLib)
print(" ")
```
